# import lib

```python
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os
from groq import Groq
from pathlib import Path
```

```python
load_dotenv()  
api_key = os.getenv("GROQ_API_KEY")
client = Groq(api_key=api_key)
 
```

```python
print("Loaded:", bool(api_key))
```


### load your local knowledge file

```python
context_text = Path("info.txt").read_text(encoding="utf-8")
print("Loaded chars:", len(context_text))
```

```python
context_text
```

```python
# optional: keep short chat history in notebook
history = []

def ask_from_info(question: str, model: str = "llama-3.3-70b-versatile"):
    messages = [
        {
            "role": "system",
            "content": (
                "Answer ONLY from the provided context. "
                "If the answer is not in context, say: "
                "'I could not find that in info.txt.'\n\n"
                f"Context:\n{context_text}"
            ),
        }
    ]
    messages.extend(history[-6:])  # keep recent turns
    messages.append({"role": "user", "content": question})

    resp = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=1
    )

    answer = resp.choices[0].message.content
    history.append({"role": "user", "content": question})
    history.append({"role": "assistant", "content": answer})
    return answer
```

```python
print(ask_from_info("What does the file say about ...?"))
```
